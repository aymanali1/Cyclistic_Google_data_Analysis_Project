```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

```python
 csv_files = glob.glob("/home/Divvy/202404-202503-divvy-tripdata/*.csv")
```

```python
csv_files.sort()
```

```python
dataframes = []
for file in csv_files:
    if os.path.exists(file):
        df = pd.read_csv(file)
        dataframes.append(df)
    else:
        print(f"file not found: {file}")
```

```python
for i, file in enumerate(csv_files):
    print(f"dataframes[{i}] ← {file}")
```

```python
dfs = dataframes
dfs[0].head()
```

```python
#Combining all dataframes into one dataframe 
#creating an empty dataframe
combined_df= pd.DataFrame()
#looping through source files
for i, file in enumerate(csv_files):
    if os.path.exists(file):
        df = pd.read_csv(file)
        #Extracting the month from filename
        month = os.path.basename(file)[:6]
        #adding a column to keep track of which file/month the data came from
        df["source_month"] = month
        #Append to the combined dataframe
        combined_df = pd.concat([combined_df, df], ignore_index=True)
    else:
        print(f"file not found: {file}")
```

```python
combined_df["started_at"] = pd.to_datetime(combined_df["started_at"], format='mixed')
combined_df["ended_at"] = pd.to_datetime(combined_df["ended_at"], format='mixed')
```

```python
combined_df["ride_length"] = combined_df["ended_at"] - combined_df["started_at"]
```

```python
# 1. Round coordinates to 4 decimal places (adjust if needed)
combined_df["start_lat_rounded"] = combined_df["start_lat"].round(4)
combined_df["start_lng_rounded"] = combined_df["start_lng"].round(4)

# 2. Build a mapping of (lat, lng) -> station name where name is not null
known_stations = combined_df.dropna(subset=["start_station_name"])
coord_to_station = dict(
    zip(
        zip(known_stations["start_lat_rounded"], known_stations["start_lng_rounded"]),
        known_stations["start_station_name"]
    )
)

# 3. Apply the mapping to fill missing start station names
def get_station_name(row):
    if pd.isna(row["start_station_name"]):
        return coord_to_station.get((row["start_lat_rounded"], row["start_lng_rounded"]), None)
    return row["start_station_name"]

combined_df["start_station_name"] = combined_df.apply(get_station_name, axis=1)
```

```python
combined_df["end_lat_rounded"] = combined_df["end_lat"].round(4)
combined_df["end_lng_rounded"] = combined_df["end_lng"].round(4)

known_end_stations = combined_df.dropna(subset=["end_station_name"])
end_coord_to_station = dict(
    zip(
        zip(known_end_stations["end_lat_rounded"], known_end_stations["end_lng_rounded"]),
        known_end_stations["end_station_name"]
    )
)

def get_end_station_name(row):
    if pd.isna(row["end_station_name"]):
        return end_coord_to_station.get((row["end_lat_rounded"], row["end_lng_rounded"]), None)
    return row["end_station_name"]

combined_df["end_station_name"] = combined_df.apply(get_end_station_name, axis=1)
```

```python
sns.countplot(data=combined_df, x='member_casual')
plt.title('Number of Rides by User Type')
plt.xlabel('User Type')
plt.ylabel('Number of Rides')
plt.show()
```

```python
# Convert ride_length to total seconds for easier averaging
combined_df['ride_length_sec'] = combined_df['ride_length'].dt.total_seconds()

# Bar plot of average ride length
avg_ride_length = combined_df.groupby('member_casual')['ride_length_sec'].mean().reset_index()

sns.barplot(data=avg_ride_length, x='member_casual', y='ride_length_sec')
plt.title('Average Ride Length by User Type')
plt.xlabel('User Type')
plt.ylabel('Avg Ride Length (seconds)')
plt.show()
```

```python
pivot_avg = combined_df.pivot_table(
    values='ride_length_sec',
    index='member_casual',
    columns='day_of_week',
    aggfunc='mean',
    observed=False
)

pivot_avg = pivot_avg[["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]]  # sort days

pivot_avg.T.plot(kind='bar', figsize=(10,6))
plt.title('Average Ride Length by Day of Week and User Type')
plt.xlabel('Day of Week')
plt.ylabel('Avg Ride Length (seconds)')
plt.xticks(rotation=45)
plt.legend(title='User Type')
plt.tight_layout()
plt.show()
```

```python
pivot_count = combined_df.pivot_table(
    values='ride_id',
    index='member_casual',
    columns='day_of_week',
    aggfunc='count',
    observed=False
)

pivot_count = pivot_count[["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]]

pivot_count.T.plot(kind='bar', figsize=(10,6))
plt.title('Number of Rides by Day of Week and User Type')
plt.xlabel('Day of Week')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.legend(title='User Type')
plt.tight_layout()
plt.show()
```
